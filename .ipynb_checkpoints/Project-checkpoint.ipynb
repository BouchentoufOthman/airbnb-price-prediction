```python
# Libraries
import pandas as pd
import numpy as np

# Load datasets
listings = pd.read_csv("listings.csv")
calendar = pd.read_csv("calendar.csv")

# Quick look
print("Listings shape:", listings.shape)
print("Calendar shape:", calendar.shape)
```

```python
# Count of available vs unavailable days
availability_counts = calendar['available'].value_counts()
print(availability_counts)

# Percentage
availability_percentage = calendar['available'].value_counts(normalize=True) * 100
print(availability_percentage)
```

```python
# Calendar price might be object like "$100.00"
# Remove $ and commas, convert to float
calendar['price'] = calendar['price'].replace('[\$,]', '', regex=True).astype(float)

# Quick check
calendar['price'].head()
```

```python
# Number of unique prices per listing
unique_prices = calendar.groupby("listing_id")["price"].nunique()

# Count dynamic vs static
dynamic_count = (unique_prices > 1).sum()
static_count = (unique_prices == 1).sum()
dynamic_percentage = (unique_prices > 1).mean() * 100

print("Dynamic listings (more than 1 price):", dynamic_count)
print("Static listings (1 price):", static_count)
print("Dynamic percentage (%):", dynamic_percentage)
```
